## 2) Splitting between Train and Test

#### before split data into test and train, we need to choose only the numerical dataset 

```python
customers.columns
```

```python
X= customers[['Avg. Session Length','Time on App','Time on Website',
          'Length of Membership']]
y= customers['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
```

## splitting into Test and Train

```python
X_train, X_test,y_train,y_test= train_test_split(X,y,test_size= 0.3,
                                                random_state=101)
```

## 3 ) building the training Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
```

```python
lm= LinearRegression()
DTmodel= DecisionTreeRegressor()
RFmodel =RandomForestRegressor()
GBmodel= GradientBoostingRegressor()
```

### fitting the model

## Non-normalized vs Normalized data

### fitting the non- normalized Data

```python
lm.fit(X_train,y_train)

```

#### fitting using the other model

```python

### fitttin
DTmodel.fit(X_train,y_train)
RFmodel.fit(X_train,y_train)
GBmodel.fit(X_train,y_train)
```

#### coefficient ( only for linear regression)

```python
print('coefficient for Linear regression : ',lm.coef_)
```

### predicting the test data

```python
prediction_lm= lm.predict(X_test)
prediction_RF= RFmodel.predict(X_test)
prediction_DT= DTmodel.predict(X_test)
prediction_GB= GBmodel.predict(X_test)
```

```python
plt.scatter(y_test,prediction_GB,label='Gradient Boosting')
plt.scatter(y_test,prediction_DT,label='Decision Tree')
plt.scatter(y_test,prediction_lm,label='Linear Regression')
plt.scatter(y_test,prediction_RF,label='Random FOrest')
# plt.show()
plt.legend(loc='best')
```

## Evaluating the model 

```python
from sklearn import metrics
```

evaluating model is as important as building model itself. it is necessary to see the performance of the model, how good our model approaches the problem and our data. because of it will be deployed on our data and will be used in real time 

calculate the 
- Mean absolute error
- Root mean square error
- Root mean absolute error 

```python
MAE= [] 
MSE= []
RMAE= []
# names= ['Mean absolute error','Root mean square error'
#         ,'Root mean absolute error']
MAE.append(metrics.mean_absolute_error(y_test,prediction_GB))
MAE.append(metrics.mean_absolute_error(y_test,prediction_lm))
MAE.append(metrics.mean_absolute_error(y_test,prediction_DT))
MAE.append(metrics.mean_absolute_error(y_test,prediction_RF))


MSE.append(metrics.mean_squared_error(y_test,prediction_GB))
MSE.append(metrics.mean_squared_error(y_test,prediction_lm))
MSE.append(metrics.mean_squared_error(y_test,prediction_DT))
MSE.append(metrics.mean_squared_error(y_test,prediction_RF))


RMAE.append(np.sqrt(metrics.mean_squared_error(y_test,prediction_GB)))
RMAE.append(np.sqrt(metrics.mean_squared_error(y_test,prediction_lm)))
RMAE.append(np.sqrt(metrics.mean_squared_error(y_test,prediction_DT)))
RMAE.append(np.sqrt(metrics.mean_squared_error(y_test,prediction_RF)))
```

```python
listmodel= ['gradient boosting ','linear regression',
           'decision Tree','Random FOrest']
evaluationmodel =pd.DataFrame(columns=['the Model used',
                      'mean absolute error',
                      'mean squared error',
                     'root mean absolute error'])
```

```python
evaluationmodel['the Model used']= listmodel
evaluationmodel['mean absolute error']= MAE
evaluationmodel['mean squared error']= MSE
evaluationmodel['root mean absolute error']= RMAE
```

```python
evaluationmodel
```

### plotting the residuals

from the residuals below we can actually evaluate how our models doing and which model yield better result (observed by the  less-wide error distribution) 

```python
sns.distplot((y_test-prediction_lm),bins=50,label='Linear regression')
sns.distplot((y_test-prediction_GB),bins=50,label='Gradient boosting')
sns.distplot((y_test-prediction_RF),bins=50,label='Random foorest')
sns.distplot((y_test-prediction_DT),bins=50,label='Decision Treee')
plt.legend(loc='best')
```
